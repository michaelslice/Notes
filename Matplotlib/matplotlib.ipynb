## Figure
**Figure**: Tracks all child axes, like titles, figures, legends, colorbars

## Axes
**Axes**: Represents actual plotting area
- x-axis, y-axis
- title, labels, grid
- plotted data itself

## Axis
**Axis**: Refers to just one of the coordinate axes, either x-axis or the y-axis 
- Each axis controls the range of values, tick marks, and labels

## Coding Styles
Two ways to use Matploblib
- Object Oriented(OO) Style: Explicitly create Figures and Axes and call methods 
- PyPlot Functions for Plotting: Rely on PyPlot to implicitly create and manage the Figures and Axes

```python
import pandas as pd
import matplotlib.pyplot as plt
from math import * 
import numpy as np

y = lambda x: x ** 2
print(y(2))

x = np.linspace(0, 2, 100) # Sample data

# .pyplot.figure to create the figure
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title("Graph Example")
ax.legend() # Add a legend
```

## Making Helper Functions
If you need to make the same plot over and over again, use this signature

```python
import pandas as pd
import matplotlib.pyplot as plt

def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100) # Make 4 random datasets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker' : 'x'})
my_plotter(ax2, data3, data4, {'marker' : 'o'})
```

## Linewidths, Linestyles, and Markersizes
- **Linewidth**: Measured in typographic points, controls thickness of lines in plot
- **Linestyle**: Defines pattern of the line
- **Markersize**: Specifies size of data points, and style

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(data1, 'o', label='data1')
ax.plot(data2, 'd', label='data2')
ax.plot(data3, 'v', label='data3')
ax.plot(data4, 's', label='data4')
ax.legend()
```

## Labeling Plots
All can be used to add text to the plot
- set_xlabel
- set_ylabel
- set_title

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# the histogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark Lengths\n (not really)')
ax.text(75, .025, r'$\mu=115, \ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

## Annotations
Annotations allow us to add text labels and arrows to highlight parts of the points in the graph

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5), 
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```

## Legends
We can specify a legend with `plt.legend()`

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

ax.plot(np.arange(len(data1)), data1, label='data1')
ax.plot(np.arange(len(data2)), data2, label='data2')
ax.plot(np.arange(len(data3)), data3, label='data3')
ax.legend()
```

## Axis Scales and Ticks
Each Axes has two to three Axis objects, that control the scale of the Axis
## Scales
Matplotlib also has non-linear scales
- loglog
- semilogx
- semilogy

```python
figs, axs = plt.subplots(1, 2, figsize=(5, 2.7), layout='constrained')
xdata = np.arange(len(data1))
data = 10 ** data1
axs[0].plot(xdata, data)

axs[1].set_yscale('log')
axs[1].plot(xdata, data)
```

## Tick Locators & Formatters
Each axis has a tick locator and a formatter that choose where along the Axis objects to put tick marks 

```python
figs, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(xdata, data1)
axs[0].set_title('Automatic Ticks')

axs[1].plot(xdata, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])
axs[1].set_title('Manual Ticks')
```

## Plotting Dates and Strings
Matplotlib can handle plotting arrays of dates and arrays of strings, these get special locators and formatters as appropriate 

```python
from matplotlib.dates import ConciseDateFormatter

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
dates = np.arange(np.datetime64('2021-11-15'), np.datetime64('2021-12-25'),
                  np.timedelta64(1, 'h'))
data = np.cumsum(np.random.randn(len(dates)))
ax.plot(dates, data)
ax.xaxis.set_major_formatter(ConciseDateFormatter(ax.xaxis.get_major_locator()))

# For strings, we get categorical plotting
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
categories = ['turnips', 'rutabaga', 'cucumber', 'pumpkins']

ax.bar(categories, np.random.rand(len(categories)))
```
